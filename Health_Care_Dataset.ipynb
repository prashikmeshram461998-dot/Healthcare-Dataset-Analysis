# Part A  Data Summarization

1. ### Load the dataset and display the first 10 rows.

```python
import pandas as pd
import numpy as np

# Load Dataset and Display First 10 Rows

df = pd.read_csv("healthcare_dataset.csv")
df.head(10)
```

### 2. Shape of the Dataset

```python
# Shape od Dataset

print("Dataset Shape : ", df.shape)

df[['Age', 'Blood_Pressure', 'BMI', 'Heart_Rate', 'Sleep_Hours']].describe()
```

### 3. Generate summary statistics for numerical columns (Age, Blood_Pressure, BMI, Heart_Rate, Sleep_Hours).

```python
# Summary Statistics for Numerical Columns

df[['Age', 'Blood_Pressure', 'BMI', 'Heart_Rate', 'Sleep_Hours']].describe
```

### 4. Calculate mean, median, mode, min, max, variance, std. deviation for Age, BMI, and Blood_Pressure.

```python
# mean, median, mode, min, max, variance, standard deviation

for col in ['Age', 'BMI', 'Blood_Pressure']:
    print(f"~~~{col}~~~")
    print(f"Mean : ", df[col].mean())
    print(f"Median : ", df[col].median())
    print(f"Mode : ", df[col].mode())
    print(f"Min : ", df[col].min())
    print(f"Max : ", df[col].max())
    print(f"Variance : ", df[col].var())
    print(f"Standard Deviation : ", df[col].std())
    print()
```

### 5. Count the number of patients in each Gender and Disease_Risk category.

```python
# Count of Patients in Each Gender and Diseas Risk Caregory

print(df['Gender'].value_counts())
print(df['Disease_Risk'].value_counts())
```

### 6. Find the distribution of Cholesterol levels.

```python
# Distribution of Cholestrol Level

print(df['Cholesterol'].value_counts(normalize = True)*100)
```

### 7. Compute the correlation matrix for numerical features.

```python
# Correlation matrix for Numerical Features.

df.corr(numeric_only=True)
```

### 8. How many patients exercise daily vs. those who never exercise?

```python
# Patients exercise Frequency

print(df['Exercise_Freq'].value_counts())
```

# Part C  Data Analysis and Insights

## 1. Visualization

### a. Histogram of Age.

```python
plt.figure(figsize = (6, 4))
sns.histplot(df['Age'], bins = 20, kde = True)
plt.title("Histogram of Age")
plt.show()
```

### b. Bar chart of Gender distribution.

```python
sns.countplot(x = 'Gender', data = df)
plt.title('Gender Distribution')
plt.show()
```

### d. Countplot of Disease_Risk.

```python
sns.countplot( x = 'Cholesterol', data = df)
plt.title('Cholesterol Levels')
plt.show()
```

### e. Boxplot of BMI across Disease_Risk.

```python
sns.countplot( x = 'Disease_Risk', data = df)
plt.title('Disease Risk Count')
```

### f. Boxplot of Blood_Pressure (Smoker vs Non-Smoker).

```python
sns.boxplot(x = 'Smoker', y = 'Blood_Pressure', data = df)
plt.title("Blood Pressure by Smoking Status")
```

### g. Pie chart of Smoker status.

```python
df['Smoker'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.title('Smoker Distribution')
plt.show()
```

### h. Stacked bar chart of Exercise_Freq vs Disease_Risk.

```python
pd.crosstab(df['Exercise_Freq'], df['Disease_Risk']).plot(kind = 'bar', stacked = True)
plt.title("ExerciseFrequency vs Disease Risk")
plt.ylabel("Count")
```

### i. Violin plot of Sleep_Hours by Gender.

```python
sns.violinplot(x = 'Gender', y = 'Sleep_Hours', data = df)
plt.title('Sleep Hours by Gender')
plt.show()
```

### j. Scatter plot of BMI vs Blood_Pressure.

```python
plt.figure(figsize = (6, 4))
sns.scatterplot(x = 'BMI', y ='Blood_Pressure', data = df, hue = 'Disease_Risk', alpha = 0.7)
plt.title('ScatterPlot : BMI vs Blood Pressure')
plt.xlabel('BMI')
plt.ylabel('Blood Pressure')
plt.show()
```

### k. Heatmap of correlations among numerical variables.

```python
plt.figure(figsize = (8, 5))
corr = df[['Age', 'BMI', 'Blood_Pressure', 'Heart_Rate', 'Sleep_Hours']].corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```

### l. Line plot of average Blood_Pressure across age groups.

```python
avg_bp_by_age = df.groupby('Age')['Blood_Pressure'].mean().reset_index()

plt.figure(figsize = (6, 4))
sns.lineplot(x = 'Age', y = 'Blood_Pressure', data = avg_bp_by_age, marker = 'o')
plt.title("Average Blood Pressure Across Age Groups")
plt.ylabel("Average Blood Pressure")
```
